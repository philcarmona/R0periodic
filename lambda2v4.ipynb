```python

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

muh=0.05
T=1
# c = ((1,0,0,0.4),(1,0,0,0.6),(1,0,0,0.8),(1,0,0,1), (0,0,1,1), (0,0,1,0.7),(0,0,1,0.5) )
alpha=0.6
colors = ( (0,0,1,alpha),(0,0,1,alpha), (0,0,1,alpha),(0,0,1,alpha),(1,0,0,alpha),(1,0,0,alpha),(1,0,0,alpha) )
cmap = mcolors.LinearSegmentedColormap.from_list('my_colormap', colors)


gamh=5
muv=1

xlist = np.linspace(0, 1, 1000)  # x=[0,1] with 1000 point
ylist = np.linspace(0, 0.1, 1000) # X x Y (1,0.1) with 1000x 1000=10^{6} pixel
x, y = np.meshgrid(xlist, ylist)
Z = (T*muv*(muh+gamh)/(0.5*(gamh+muh+muv)))*(x*y-( ((x+y)*muv-y*(gamh+muh))*( (x+y)*(gamh+muh)-y*muv )) /(4*np.pi**2/T**2+(gamh+muh+muv)**2))
plt.figure()


# levels = [-0.15,-0.125,-0.1,-0.075,-0.05,-0.025,0    ,0.025]
levels = [-0.075,-0.05,-0.025,0,0.025,0.05,0.075]
# levels = [-0.09,-0.06,-0.03,0,0.03,0.06,0.09]

fig, ax=plt.subplots(figsize=(4,3))

# contour = plt.contour(x, y, Z)
# plt.clabel(contour, colors = 'k',fontsize=8)
# contour_filled = plt.contourf(x, y, Z,colors=c, linewidths=0.1)

# Create filled contour plot
cf = ax.contourf(x, y, Z, levels=[-100,0.0,100], cmap=cmap, zorder=0)  # 8 is the number of colors we want to generate to colouring domain

# Create contour plot with black isolines
cntr = ax.contour(x, y, Z, levels=levels, colors='black', linewidths=0.5,zorder=1)

# Label contour lines
# ax.clabel(cntr, inline=1,inline_spacing=10, fontsize=10)

# Define manual label
label_locations = [(0.9, 0.01),(0.8, 0.02),(0.6, 0.02),(0.5, 0.02),(0.5, 0.06),(0.6, 0.08),(0.8, 0.09)]
labels = ax.clabel(cntr, inline=True, fontsize=10, manual=label_locations)


for c in cntr.collections:
    c.set_linestyle('solid')

plt.title('$T=1$',fontsize=12)
plt.xlabel('$C_{I}$',fontsize=12)
plt.ylabel('$C_{II}$',fontsize=12)
plt.savefig('P1secondoderT1.pdf',bbox_inches = 'tight')
plt.show()
```
